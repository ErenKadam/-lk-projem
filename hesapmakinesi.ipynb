```python
import numpy as np
def hesap_makinesi():
    while True:
        print("HESAP MAKİNESİ")
        print("1. Toplama")
        print("2. Çıkarma")
        print("3. Çarpma")
        print("4. Bölme")
        print("5. Üs alma")
        print("6. Karekök alma")
        print("0. Çıkış")

        secim = input("Seçiminizi yapın: ")

        if secim == "0":
            print("Çıkış yapılıyor...")
            break

        elif secim in ["1","2","3","4","5"]:
            sayi1 = float(input("Birinci sayıyı girin: "))
            sayi2 = float(input("İkinci sayıyı girin: "))

            if secim == "1":
                sonuc = np.add(sayi1, sayi2)
            elif secim == "2":
                sonuc = np.subtract(sayi1, sayi2)
            elif secim == "3":
                sonuc = np.multiply(sayi1, sayi2)
            elif secim == "4":
                if sayi2 != 0:
                    sonuc = np.divide(sayi1, sayi2)
                else:
                    sonuc = "Hata! Sıfıra bölünemez."
            elif secim == "5":
                sonuc = np.power(sayi1, sayi2)

            print(f"Sonuç: {sonuc}")

        elif secim == "6":
            sayi = float(input("Bir sayı girin: "))
            if sayi >= 0:
                sonuc = np.sqrt(sayi)
            else:
                sonuc = "Hata! Negatif sayının karekökü alınamaz."
            print(f"Sonuç: {sonuc}")

        else:
            print("Geçersiz seçim, lütfen tekrar deneyin.")
hesap_makinesi()
```
