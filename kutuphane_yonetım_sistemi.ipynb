```python
# Kütüphane Yönetim Sistemi

kutuphane = []  # kitapları saklayacağımız liste

def kitap_ekle():
    isim = input("Kitap adı: ")
    yazar = input("Yazar: ")
    yil = input("Yıl: ")
    kitap = {"isim": isim, "yazar": yazar, "yil": yil}
    kutuphane.append(kitap)
    print(f"'{isim}' kütüphaneye eklendi.\n")

def kitap_sil():
    isim = input("Silmek istediğiniz kitabın adı: ")
    for kitap in kutuphane:
        if kitap["isim"].lower() == isim.lower():
            kutuphane.remove(kitap)
            print(f"'{isim}' kütüphaneden silindi.\n")
            return
    print("Kitap bulunamadı.\n")

def kitaplari_listele():
    if not kutuphane:
        print("Kütüphane boş.\n")
    else:
        print("--- KÜTÜPHANEDEKİ KİTAPLAR ---")
        for i, kitap in enumerate(kutuphane, 1):
            print(f"{i}. {kitap['isim']} - {kitap['yazar']} ({kitap['yil']})")
        print()

def kitap_ara():
    kelime = input("Aramak istediğiniz kitap adı veya yazar: ")
    bulundu = False
    for kitap in kutuphane:
        if kelime.lower() in kitap["isim"].lower() or kelime.lower() in kitap["yazar"].lower():
            print(f"{kitap['isim']} - {kitap['yazar']} ({kitap['yil']})")
            bulundu = True
    if not bulundu:
        print("Aradığınız kitap bulunamadı.\n")

# Ana döngü
while True:
    print("""
    --- KÜTÜPHANE YÖNETİM SİSTEMİ ---
    1. Kitap Ekle
    2. Kitap Sil
    3. Kitapları Listele
    4. Kitap Ara
    5. Çıkış
    """)
    
    secim = input("Seçiminizi yapın: ")

    if secim == "1":
        kitap_ekle()
    elif secim == "2":
        kitap_sil()
    elif secim == "3":
        kitaplari_listele()
    elif secim == "4":
        kitap_ara()
    elif secim == "5":
        print("Programdan çıkılıyor...")
        break
    else:
        print("Geçersiz seçim. Lütfen tekrar deneyin.\n")
```
